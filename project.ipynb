```python
import requests
import bs4
import urllib
import datetime
import openpyxl as op
#'%Y-%m-%d'
```

# Extracting north indian food and its offers

```python
s=open('D:/study/aegis/python/python-proj-release-1.0/data/swiggy/Northindian9-7-19.txt')
soup=bs4.BeautifulSoup(s,'html.parser')
cuisine=[]
restaurant=[]
offers=[]
date=[]
wb = op.load_workbook('Swiggy.xlsx')
ws = wb.get_sheet_by_name('sheet1')
for line in soup:
    f=line.findAll('div',{'class':'_3FR5S'})
    for i in f:
        if i.find('div',{'class':'Zlfdx'})!=None:
            a1=i.findAll('div',{'class':'_3Ztcd'})
            for j in a1:
                restaurant.append(j.find('div',{'class':'nA6kb'}).text)
for line in soup:
    b1=line.findAll('div',{'class':'Zlfdx'})
    for i in b1:
        b2=i.get_text('span',{'class':'sNAfh'})
        offers.append(b2)
        date.append(datetime.datetime.today().strftime('%Y-%m-%d'))
        cuisine.append('North Indian')
o=list(zip(cuisine,restaurant,offers,date))
for i in o:
    ws.append(i)
wb.save('Swiggy.xlsx')
wb.close()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.DataFrame(dict(
    online=[90,97,117,94,76,90],
    dine=[293,308,328,309,293,239],
    sw_offers=[149,74,114,102,82,124]))
fig = plt.figure(figsize=(20, 10))
index=['25th june','26th june','27th june','28th june','2nd july','9th july']
z_bar=[plt.bar(index,df.dine,align='edge',width=0.2,label='Zomato dine'),
       plt.bar(index,df.online,align='edge',width=0.2,label='Zomato online')]
s_bar=plt.bar(index,df.sw_offers,align='edge',width=-0.2,label='Swiggy offers')
plt.xlabel('Dates',fontsize=16)
plt.ylabel('Number of offers',fontsize=16)
plt.title('Date wise offers in Swiggy and Zomato',fontsize=16)
plt.legend(loc='upper left')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
N=6
ind=np.arange(N)
dates=['25th june','26th june','27th june','28th june','2nd July','9th July']
onlinez=[90,97,117,94,76,90]
onlines=[149,74,114,102,82,124]
fig = plt.figure(figsize=(20, 10))
Type=['Zomato online','Swiggy Online']
plt.bar(ind,onlinez,align='center',width=0.2)
plt.bar(ind+0.2,onlines,align='center',width=0.2)
plt.xticks(ind,dates)
plt.xlabel('Dates',fontsize=16)
plt.ylabel('Number of offers',fontsize=16)
plt.title('Online offers in Swiggy and Zomato',fontsize=16)
plt.legend(Type,loc=2)
plt.show()
```
