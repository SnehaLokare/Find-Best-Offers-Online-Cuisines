```python
df = df_dup['Offer'].str.split(',')
dine = []
online = []
for d in df:
    for i in d:
        if ('gold' in i or 'dine' in i) or ('order' not in ''.join(d)):
            if ('order' not in ''.join(d)):
                dine.append(','.join(d))
                online.append([])
                break
            dine.append(i)
            if ('order' not in i and len(d) == 1): 
                online.append([])
            
        if 'order' in i:
            online.append(i)
            if 'dine' not in i and len(d) == 1:
                dine.append([])
len(online),len(dine),len(df_dup['Offer'])
# online[160:180]
df_dup['online'] = online
df_dup['dine'] = dine
len(dine),len(online)
```

```python
def get_dine_offer(dine):
    if(dine):
#       print(dine)
        find = re.findall('\d[+]\d',dine)
#         print(dine,find)
        result = 0
        if(find):
            for i in find:
                if i == '1+1':
                    result+=50
                elif i == '2+2':
                    result+=25
            return result
        return dine
    return 0
```

```python
import re
import numpy as np
temp = df_dup['dine'].apply(get_dine_offer)
df_dup['dine'] = temp
```

```python
ds = pd.read_excel('Swiggy.xlsx',sheet_name = 'sheet4')
dsw = ds[ds['DATE']=='2019-07-02' ]
# print(dsw)
x=dsw.groupby(['CUISINES'])['OFFERS'].count()
print(x)
dfz=pd.read_excel('zomato3.xlsx',sheet_name='final')
dz = dfz[dfz['Date']=='02-07-2019' ]
y=dz.groupby(['Cuisines'])['online'].agg(np.count_nonzero)
z=dz.groupby(['Cuisines'])['dine'].agg(np.count_nonzero)
print(y,z)
```
